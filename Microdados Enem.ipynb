# Microdados Enem

```python
# Importando todas as biblioteca

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Criando dataframes das bases de dados dos anos 2022 a 2019 e trazendo apenas as colunas que vão ser utilizadas

enem_2022 = pd.read_csv('Arquivos/MICRODADOS_ENEM_2022.csv', sep=';', encoding='ISO-8859-1', usecols=['NU_INSCRICAO','NU_ANO','CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA','SG_UF_PROVA','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO','Q024'])

enem_2021 = pd.read_csv('Arquivos/MICRODADOS_ENEM_2021.csv', sep=';', encoding='ISO-8859-1', usecols=['NU_INSCRICAO','NU_ANO','CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA','SG_UF_PROVA','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO','Q024'])

enem_2020 = pd.read_csv('Arquivos/MICRODADOS_ENEM_2020.csv', sep=';', encoding='ISO-8859-1', usecols=['NU_INSCRICAO','NU_ANO','CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA','SG_UF_PROVA','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO','Q024'])

enem_2019 = pd.read_csv('Arquivos/MICRODADOS_ENEM_2019.csv', sep=';', encoding='ISO-8859-1', usecols=['NU_INSCRICAO','NU_ANO','CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA','SG_UF_PROVA','NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO','Q024'])
```

```python
# Visualizando as os dataframes

display(enem_2022)
display(enem_2021)
display(enem_2020)
display(enem_2019)
```

# Criando uma figura com 4 gráficos de barras das médias de cada máterias por região

```python
# Tirando todas as linhas que tenham pelo menos um valor nulo

regiao_2022 = enem_2022.dropna()
regiao_2021 = enem_2021.dropna()
regiao_2020 = enem_2020.dropna()
regiao_2019 = enem_2019.dropna()
```

```python
# Dicionario contendo todas as regiões do brasil com sigla dos estados

regioes = {'Norte':['AM','RR','AP','PA','TO','RO','AC'],
           'Nordeste':['MA','PI','CE','RN','PE','PB','SE','AL','BA'],
           'Centro-Oeste':['MT','MS','GO','DF'],
           'Sudeste':['SP','RJ','ES','MG'],
           'Sul':['PR','RS','SC']}

# Crio uma função para verificar qual região é correta para cada estado

def mapear_regiao(estado):
    for regiao, estados_regiao in regioes.items():
        if estado in estados_regiao:
            return regiao
    return 'Região desconhecida'
```

```python
# Adicionando a coluna de região usando a função acima para todos os dataframes

regiao_2022.insert(loc=5,column='REGIAO', value=regiao_2022['SG_UF_PROVA'].map(mapear_regiao))
regiao_2021.insert(loc=5,column='REGIAO', value=regiao_2021['SG_UF_PROVA'].map(mapear_regiao))
regiao_2020.insert(loc=5,column='REGIAO', value=regiao_2020['SG_UF_PROVA'].map(mapear_regiao))
regiao_2019.insert(loc=5,column='REGIAO', value=regiao_2019['SG_UF_PROVA'].map(mapear_regiao))
```

```python
# Visualizando a nova coluna 

display(regiao_2022.head())
display(regiao_2021.head())
display(regiao_2020.head())
display(regiao_2019.head())
```

```python
# Crio uma nova váriavel para contem uma dataframe com as notas média de cada região por materia

media_regiao_2022 = regiao_2022.groupby('REGIAO').mean(numeric_only=True).drop(['NU_INSCRICAO','CO_MUNICIPIO_PROVA'],axis=1)
media_regiao_2021 = regiao_2021.groupby('REGIAO').mean(numeric_only=True).drop(['NU_INSCRICAO','CO_MUNICIPIO_PROVA'],axis=1)
media_regiao_2020 = regiao_2020.groupby('REGIAO').mean(numeric_only=True).drop(['NU_INSCRICAO','CO_MUNICIPIO_PROVA'],axis=1)
media_regiao_2019 = regiao_2019.groupby('REGIAO').mean(numeric_only=True).drop(['NU_INSCRICAO','CO_MUNICIPIO_PROVA'],axis=1)
```

```python
# Visualizando as medias por região do brasil

display(media_regiao_2022)
display(media_regiao_2021)
display(media_regiao_2020)
display(media_regiao_2019)
```

```python
# Defina as informações necessárias para o gráficos
regioes = media_regiao_2022.index
colunas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']
num_regioes = len(regioes)
num_colunas = len(colunas)

# Crie um array com as médias de todas as notas para cada região
medias_notas_2022 = media_regiao_2022[colunas].values
medias_notas_2021 = media_regiao_2021[colunas].values
medias_notas_2020 = media_regiao_2020[colunas].values
medias_notas_2019 = media_regiao_2019[colunas].values

# Crie uma sequência de valores para o eixo x
posicao_barras = np.arange(num_regioes)

# Defina a largura das barras agrupadas
largura_barras = 0.15

# Crie uma figura que vai ter 4 graficos de barras agrupadas com as médias de todas as notas
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 8))

for i in range(num_colunas):
    ax[0,0].bar(posicao_barras + (i * largura_barras), medias_notas_2022[:, i], largura_barras)
    
for i in range(num_colunas):
    ax[0,1].bar(posicao_barras + (i * largura_barras), medias_notas_2021[:, i], largura_barras)
    
for i in range(num_colunas):
    ax[1,0].bar(posicao_barras + (i * largura_barras), medias_notas_2020[:, i], largura_barras)

for i in range(num_colunas):
    ax[1,1].bar(posicao_barras + (i * largura_barras), medias_notas_2019[:, i], largura_barras)
    

# Define o rótulo do eixo y, e adiciona títulos aos gráficos
ax[0,0].set_ylabel('Média das notas')
ax[0,0].set_title('Média das notas por região 2022')
ax[0,1].set_title('Média das notas por região 2021')
ax[1,0].set_title('Média das notas por região 2020')
ax[1,1].set_title('Média das notas por região 2019')

# Adicione uma legenda a figura
colunas_personalizadas = ['Ciências da Natureza', 'Ciências Humanas', 'Linguagens e Códigos', 'Matemática', 'Redação']
fig.legend(colunas_personalizadas,loc='upper center',bbox_to_anchor=(0.512, 1.01),fontsize='small').get_frame().set_alpha(0)

# Adiciona uma grade para todos os gráficos e alguns configuraçoes no ticks
for row in ax:
    for axes in row:
        axes.grid(axis='y',linestyle = "dashed")
        axes.set_xticks([0.1, 1.3 ,2.3, 3.3, 4.3])
        axes.set_xticklabels(regioes)
        axes.set_yticks([0,100,200,300,400,500,600,650,700])
        axes.set_ylim(0,800)

# Exiba o gráfico
plt.show()
```

# Criando um Gráfico de dispersão para cada ano (2022 a 2019) das médias de cada municipio do brasil

```python
# Tirando todas as linhas que tenham pelo menos um valor nulo

municipios_2022 = enem_2022.dropna()
municipios_2021 = enem_2021.dropna()
municipios_2020 = enem_2020.dropna()
municipios_2019 = enem_2019.dropna()
```

```python
# Faço a média das notas para cada município nos 4 ultimos anos

media_municipios_2022 = municipios_2022.groupby(['CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA'],as_index=False).mean(numeric_only=True).drop(['NU_INSCRICAO'],axis=1)
media_municipios_2021 = municipios_2021.groupby(['CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA'],as_index=False).mean(numeric_only=True).drop(['NU_INSCRICAO'],axis=1)
media_municipios_2020 = municipios_2020.groupby(['CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA'],as_index=False).mean(numeric_only=True).drop(['NU_INSCRICAO'],axis=1)
media_municipios_2019 = municipios_2019.groupby(['CO_MUNICIPIO_PROVA','NO_MUNICIPIO_PROVA'],as_index=False).mean(numeric_only=True).drop(['NU_INSCRICAO'],axis=1)
```

```python
# Visualizo as médias dos municípios

display(media_municipios_2022)
display(media_municipios_2021)
display(media_municipios_2020)
display(media_municipios_2019)
```

```python
# importo uma base de dados com longitude e latitude dos de cada município do brasil

municipios_df = pd.read_csv('Arquivos/municipios.csv',usecols=['codigo_ibge','latitude','longitude'])
```

```python
# Visualizando a base importada

display(municipios_df)
```

```python
# Mesclo a base de dados com latitude e longitude junto com as das médias

media_municipios_2022 = pd.merge(media_municipios_2022, municipios_df, how='left', left_on='CO_MUNICIPIO_PROVA', right_on='codigo_ibge')
media_municipios_2021 = pd.merge(media_municipios_2021, municipios_df, how='left', left_on='CO_MUNICIPIO_PROVA', right_on='codigo_ibge')
media_municipios_2020 = pd.merge(media_municipios_2020, municipios_df, how='left', left_on='CO_MUNICIPIO_PROVA', right_on='codigo_ibge')
media_municipios_2019 = pd.merge(media_municipios_2019, municipios_df, how='left', left_on='CO_MUNICIPIO_PROVA', right_on='codigo_ibge')
```

```python
# Visualizo se as bases foram mescladas corretamente

display(media_municipios_2022)
display(media_municipios_2021)
display(media_municipios_2020)
display(media_municipios_2019)
```

```python
# Configurando o gráfico e visualizando

figu, axe = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True)

figu.suptitle('Média da Nota de Redação nos Municípios')

vmax, vmin = 700, 350

media_municipios_2022.plot.scatter(y='latitude', x='longitude', title='Ano de 2022', figsize=(10,8), c=media_municipios_2022['NU_NOTA_REDACAO'], cmap='RdYlGn', ax=axe[0,0], vmax=vmax, vmin=vmin)
media_municipios_2021.plot.scatter(y='latitude', x='longitude', title='Ano de 2021', figsize=(10,8), c=media_municipios_2021['NU_NOTA_REDACAO'], cmap='RdYlGn', ax=axe[0,1], vmax=vmax, vmin=vmin)
media_municipios_2020.plot.scatter(y='latitude', x='longitude', title='Ano de 2020', figsize=(10,8), c=media_municipios_2020['NU_NOTA_REDACAO'], cmap='RdYlGn', ax=axe[1,0], vmax=vmax, vmin=vmin)
media_municipios_2019.plot.scatter(y='latitude', x='longitude', title='Ano de 2019', figsize=(10,8), c=media_municipios_2019['NU_NOTA_REDACAO'], cmap='RdYlGn', ax=axe[1,1], vmax=vmax, vmin=vmin)

plt.show()
```
